```python
import pandas as pd
from pycaret.regression import *
```

```python
dir = '/content/drive/MyDrive/Dacon/Iris/Dataset/'
train = pd.read_csv(dir + 'iris_train.csv')
test = pd.read_csv(dir + 'iris_test.csv')
sub = pd.read_csv(dir + 'sample_submission.csv')
```

```python
train_petal, train_sepal = train.drop(columns = ['id', 'sepal width (cm)']), train.drop(columns = ['id', 'petal width (cm)'])
test = test.drop(columns = ['id'])
```

```python
setup_sepal = setup(train_sepal, target = 'sepal width (cm)',
                    normalize = True, normalize_method = 'robust')
```

```python
top3 = compare_models(sort = 'MAE', n_select = 3)
```

```python
lr = create_model('lr', cross_validation = False)
```

```python
#tuned_top3 = [tune_model(i, optimize = 'MAE') for i in top3]
tuned_lr = tune_model(lr, optimize = 'MAE')
```

```python
#blender_top3 = blend_models(tuned_top3, optimize = 'MAE')
```

```python
final_model = finalize_model(tuned_lr)
sub['sepal width (cm)'] = predict_model(final_model, data=test).Label;
```

```python
setup_petal = setup(train_petal, target = 'petal width (cm)',
                    normalize = True, normalize_method = 'robust')
```

```python
top3 = compare_models(sort = 'MAE', n_select = 3)
```

```python
lr = create_model('lr', cross_validation = False)
```

```python
#tuned_top3 = [tune_model(i, optimize = 'MAE') for i in top3]
tuned_lr = tune_model(lr, optimize = 'MAE')
```

```python
#blender_top3 = blend_models(tuned_top3, optimize = 'MAE')
```

```python
final_model = finalize_model(tuned_lr)
sub['petal width (cm)'] = predict_model(final_model, data=test).Label;
```

```python
sub.to_csv(dir + 'submission_211029.csv', index=False)
```
